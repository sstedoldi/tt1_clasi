## Python venv on Windows

Using PowerShell

1️⃣ Create the venv (using the Windows Python launcher)

> py -3 -m venv venv # 3.11

2️⃣ Activate it

If you get an execution‐policy error, you can temporarily bypass it:

Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass

> .\venv\Scripts\Activate.ps1

3️⃣ Upgrade pip

> python -m pip install --upgrade pip

4️⃣ Install your requirements

> pip install -r requirements.txt

Running manual pip install for torch for laptop GPU

> pip install torch==2.5.1 --index-url https://download.pytorch.org/whl/cu121

---

Re-instalación por problemas con el GPU

> pip install --force-reinstall torch==2.5.1 torchvision==0.20.1 torchaudio==2.5.1 --index-url https://download.pytorch.org/whl/cu121
